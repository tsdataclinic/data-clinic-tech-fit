# Geospatial Visualization

```python
gdf = gpd.read_file("data/processed/area_median_income_dataset_pumas_ma.geojson",drive="GeoJSON")
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

gdf.plot(
    column="estimated_area_median_income_householder_age_65_and_over",
    figsize=(12, 12),
    alpha=0.5,
    edgecolor='k',
    legend=True,
    ax=ax
)

#leg = ax.get_legend(title="Area Median Income")
ax.set(title='Area Median Income for Householders Aged 65 and Over (ACS 2020 5-year)')

ax.legend(
    loc=(1, .5),
    fontsize=15,
    frameon=False
)

ax.set_axis_off()

#cx.add_basemap(ax, crs='EPSG:4326', source=cx.providers.Stamen.TonerLite)
#cx.add_basemap(ax, crs='EPSG:4326', source=cx.providers.Stamen.TonerLabels)
```
